## 循环结构4大要素
- 初始的赋值变量
- 判断语句
- 程序执行的代码体
- 可迭代的变量

```python
# 求1~100的和
a=1
sum_a=0
while a<=100:
    sum_a=sum_a+a
    a=a+1
else:
    print(sum_a)    
```

```python
# 打印乘法表
j=1
i=1
while j<=10:
    while i<=j:
        print(f"{i}×{j}={i*j}", end="\t") # t为制表符
        i=i+1
    print() #做换行
    j=j+1
    i=1
    

```

```python
i=1
while i<=6:
    print("*"*i)
    i+=1
```

```python
#根据用户成绩进行评级
while True:
    score=int(input("请输入您的成绩："))
    if score<0 or score>100:
        print("请输入一个有效成绩！")
        continue
    if score>=90:
        print("成绩评定为A")
    elif score>=80:
        print("成绩评定为B")
    else:
        print("成绩评定为C")
    break # 结束循环，程序执行完毕
```

```python
# for循环（左闭右开）
for i in range(10):
    print(i)
```

```python
# for循环打印乘法表
for j in range(11):
    i=1
    while i<=j:
        print(f"{i}×{j}={i*j}",end="\t")
        i+=1
    print()
```

## 递归：
递归指在一个函数体内调用该函数本身

```python
# 使用递归函数计算阶乘
def multi(n):
    if n==1:
        result=n
        return result
    else:
        result=n*multi(n-1)
        return result
multi(5)
```

```python
# 求解100~1000以内的水杨花数
count=0
for num in range(100,1000):
    a=num//100
    b=num//10%10
    c=num%10
    if num==a**3+b**3+c**3:
        print(num)
        count+=1
print(f"一共有{count}个水杨花数")
```

```python
def count_narcissistic_numbers(start, end):
    count = 0
    results = []

    # 遍历范围内的每一个数
    for num in range(start, end):
        # 提取百位、十位和个位
        hundreds = num // 100
        tens = (num // 10) % 10
        units = num % 10
        
        # 计算各位数字的立方和
        # 公式：a³ + b³ + c³ = num
        if hundreds**3 + tens**3 + units**3 == num:
            results.append(num)
            count += 1
            
    return count, results

# 执行计算
lower, upper = 100, 1000
total_count, found_numbers = count_narcissistic_numbers(lower, upper)

# 输出结果
print(f"在 {lower} 到 {upper} 范围内：")
print(f"找到的水仙花数有：{found_numbers}")
print(f"**总个数为：{total_count} 个**")
```
